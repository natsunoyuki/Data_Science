```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# Raw data downloaded from: https://data.gov.sg/dataset/resale-flat-prices
# The raw data comes in several csv files. Combine those csv files into
# a single consolidated csv.zip file.

base_path = "./"

Data = []

cols = []
for f in os.listdir(base_path):
    if f[:18] == "resale-flat-prices" and f[-3:] == "csv":
        print(f)
        path = base_path + f
        data = pd.read_csv(path)
        print('Loaded data shape:', data.shape)
        print("Loaded data columns:", data.columns)
        Data.append(data)
        cols.append(data.columns)
```

```python
cols_to_keep = ['month', 'town', 'flat_type', 'block', 'street_name', 'storey_range', 'floor_area_sqm', 
                'flat_model', 'lease_commence_date', 'resale_price']

data = pd.DataFrame()
for dat in Data:
    dat = dat[cols_to_keep]
    display(dat.head(1))
    data = pd.concat([data, dat], axis = 0)
```

```python
print("Final merged shape:",data.shape)
display(data.head())
```

```python
zip_file = "consolidated-resale-flat-prices.csv.zip"

data.to_csv(zip_file, index = False, compression = "zip")
```
