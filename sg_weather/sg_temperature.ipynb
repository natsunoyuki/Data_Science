```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import mode

base_dir = "sg_weather/"

# Data downloaded from: https://data.gov.sg/search?groups=environment
data = pd.read_csv(base_dir + "surface-air-temperature-monthly-mean.csv")

data['month'] = pd.DatetimeIndex(data['month'])

display(data)
```

```python
plt.figure(figsize = (15, 5))
plt.plot(data['month'], data['mean_temp'])
plt.show()
```
