```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)

pi = np.pi
```

```python
n_rows = 10000
n_iters = 5

R = np.zeros(n_rows)

for i in range(n_iters):
    R = R + np.random.exponential(1.0, 10000)
R = R / n_iters

# Calculate the analytical normal distributions
x = np.linspace(0, 3, 1000)

# Analytical mean of the normal distribution is simply 1
mu = 1 ** -1  
# Variance depends on the number of samples drawn
sig2 = 1.0 / (n_iters * 1 ** 2)  
p = ((2 * pi * sig2)**-0.5) * np.exp(-(x - mu) ** 2 / 2 / sig2)

plt.figure(figsize = (15, 5))
plt.title("Average over {} iterations".format(n_iters))
plt.hist(R, 100, density = True)
plt.plot(x, p)
plt.grid(True)
plt.xlabel("x")
plt.ylabel("p(x)")
plt.show()
```
